```python
import pandas as pd
import os
import PIL
from PIL import Image
```

```python
filter = PIL.Image.BILINEAR

for file in os.listdir('assets/Images'):
    if 'rect' in file:
        print(file)
        im = Image.open(os.path.join('assets/Images',file))
        img = im.thumbnail((256, 256), filter)
        img = im
        q = file.split('.')[0]

        img.save('assets/Images/'+q+'_thumb.png', "PNG", quality=100)
```
